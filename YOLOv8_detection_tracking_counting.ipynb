# YOLOv8     Cars object detection , tracking and counting


```python
#importing libraries
import cv2
from PIL import Image
import pandas as pd
import cvzone
```

## Importing pretrained YOLOv8

```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt") 
```

# YOLOv8 detection tracking and counting with opencv

```python
cy1=383
counter=[]
count=0
offset=4
video_path = "12.mp4"
cap = cv2.VideoCapture(video_path)
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
   
size = (frame_width, frame_height)
   
# Below VideoWriter object will create
# a frame of above defined The output 
# is stored in 'filename.avi' file.
result = cv2.VideoWriter('filename2.avi', 
                         cv2.VideoWriter_fourcc(*'MJPG'),
                         20, size)
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()
    

    if success:
        # Run YOLOv8 inference on the frame
        results = model.track(frame, conf=0.2,persist=True,classes=2)
        a=results[0].cpu().boxes.boxes
        px=pd.DataFrame(a).astype("float")
        list=[]

        for index,row in px.iterrows():
    #        print(row)

            x1=int(row[0])
            y1=int(row[1])
            x2=int(row[2])
            y2=int(row[3])
            id=int(row[4])
            
            cx=int(x1+x2)//2
            cy=int(y1+y2)//2
            
            
            
            if cy1<(cy+offset) and cy1 > (cy-offset):
                if counter.count(id)==0:
                    counter.append(id)
                    cv2.circle(frame,(cx,cy),4,(0,0,255),-1)
                    
        cv2.line(frame ,(335,cy1),(1000,cy1),(0,255,0),2)
        u=(len(counter))
        cvzone.putTextRect(frame, f'cars count:{u}', (50,60),2,2)
        frame = results[0].plot()
        result.write(frame)
        cv2.imshow("RGB", frame)



        # Break the loop if 'q' is pressed
        if cv2.waitKey(30) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reachedq
        break

# Release the video capture object and close the display window
cap.release()
result.release()
cv2.destroyAllWindows()
```
